### Using BytesIO

```python
import cv2
import numpy as np
from IPython.display import Image, display
import io
import PIL.Image
```

```python
def showarray(a, fmt='jpeg'):
    a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    a = np.uint8(np.clip(a, 0, 255))
    f = io.BytesIO()
    PIL.Image.fromarray(a).save(f, fmt)
    display(Image(data=f.getvalue()))
```

```python
img = cv2.imread('assets/01.jpg')
```

```python
showarray(img)
```

### Using matplotlib

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_img(a):
    a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    plt.imshow(a)
    plt.axis('off')
    return
```

```python
show_img(img)
```
